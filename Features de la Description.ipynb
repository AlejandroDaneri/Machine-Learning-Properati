```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
properati = pd.read_csv('datos/properati_final.csv')
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:97% !important; }</style>"))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
properati.dropna(subset=["description"],inplace=True)
properati.info()

contador = CountVectorizer()
contador.fit(properati["description"])
```

```python
X=contador.transform(properati["description"])

from sklearn.decomposition import TruncatedSVD
componentes=10
svd=TruncatedSVD(n_components=componentes)

X_svd=(svd.fit_transform(X))

properati["features"]= [np.split(row,componentes) for row in X_svd]

print nombres

properati[nombres]=pd.DataFrame(properati.features.values.tolist())
```
