```python
from sklearn.feature_extraction.text import CountVectorizer
properati.dropna(subset=["description"],inplace=True)
properati.info()

contador = CountVectorizer()
contador.fit(properati["description"])
```

### SVD

```python
X=contador.transform(properati["description"])

from sklearn.decomposition import TruncatedSVD
componentes=10
svd=TruncatedSVD(n_components=componentes)

X_svd=(svd.fit_transform(X))

properati["features"]= [np.split(row,componentes) for row in X_svd]

print nombres

properati[nombres]=pd.DataFrame(properati.features.values.tolist())
```

### Busco features

```python
properati = pd.read_csv('datos/properati_final.csv')
properati.dropna(subset=["description"],inplace=True)
```

```python
descripcion = properati["description"]
```

```python
import re
palabras=[]
for row in descripcion:
    row =row.lower()
    row=filter(None, re.split("[, \-!?:._¡¿'`<>;()+]", row))
    for palabra in row:
        if len(palabra)<4: #saco conectores
            continue
        palabras.append(palabra)
```

```python
from collections import Counter
contador = Counter(palabras)
```

```python
contador.most_common(500)
```

```python
contador.most_common()[-300:]   
```

```python
features_nuevas=["cochera","comedor","living","amplio","suite","lavadero","cochera","terraza","hall","placard","balcon","parrilla","aire","patio","muebles","pileta"\
                ,"mesada","parque","quincho","escalera","chalet","monoambiente","porcelanato","caldera","solarium","duplex","cancha","ascensor","recepción","vigilancia"]
```

```python
for feature in features_nuevas:
    properati.loc[:,feature]=descripcion.str.lower().str.contains(feature)
```

```python
properati.to_csv('datos/properati_final_y_descripcion.csv',index=False)
```
